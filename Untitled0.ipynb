### Actividad 1: Juego de adivinar el número

Aquí está el código para el juego de adivinar el número, implementando todos los pasos solicitados.

```python
import random

def generar_numero():
    """Retorna un número aleatorio entre 1 y 100."""
    return random.randint(1, 100)

numero_secreto = generar_numero()
intentos = 0

print("¡Bienvenido al Juego de Adivinar el Número!")
print("He generado un número entre 1 y 100. ¡Intenta adivinarlo!")
```

```python
while True:
    try:
        entrada_usuario = input("Introduce tu adivinanza: ")
        adivinanza = int(entrada_usuario)
        intentos += 1

        if adivinanza < numero_secreto:
            print("¡Demasiado bajo! Intenta de nuevo.")
        elif adivinanza > numero_secreto:
            print("¡Demasiado alto! Intenta de nuevo.")
        else:
            print(f"¡Felicidades! Has adivinado el número {numero_secreto} en {intentos} intentos.")
            break
    except ValueError:
        print("Entrada no válida. Por favor, introduce un número entero.")
```

### Actividad 2: Calculadora de operaciones básicas

Aquí tienes el código para la calculadora, implementando las funciones, el menú y el manejo de errores.

```python
def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def multiplicacion(a, b):
    return a * b

def division(a, b):
    if b == 0:
        raise ValueError("No se puede dividir por cero.")
    return a / b

print("¡Bienvenido a la Calculadora Básica!")
```

```python
while True:
    print("\nSelecciona una operación:")
    print("1. Suma")
    print("2. Resta")
    print("3. Multiplicación")
    print("4. División")
    print("5. Salir")

    opcion = input("Introduce tu opción (1/2/3/4/5): ")

    if opcion == '5':
        print("¡Hasta luego!")
        break

    if opcion in ('1', '2', '3', '4'):
        try:
            num1 = float(input("Introduce el primer número: "))
            num2 = float(input("Introduce el segundo número: "))
        except ValueError:
            print("Entrada no válida. Por favor, introduce números.")
            continue

        if opcion == '1':
            print(f"Resultado: {num1} + {num2} = {suma(num1, num2)}")
        elif opcion == '2':
            print(f"Resultado: {num1} - {num2} = {resta(num1, num2)}")
        elif opcion == '3':
            print(f"Resultado: {num1} * {num2} = {multiplicacion(num1, num2)}")
        elif opcion == '4':
            try:
                print(f"Resultado: {num1} / {num2} = {division(num1, num2)}")
            except ValueError as e:
                print(f"Error: {e}")
    else:
        print("Opción no válida. Por favor, intenta de nuevo.")

    input("Presiona Enter para continuar...") # Pausa para que el usuario vea el resultado
```

### Actividad 3: Lista de tareas

Aquí está el código para la lista de tareas, implementando todas las funcionalidades solicitadas.

```python
tareas = []

def agregar_tarea(tarea):
    tareas.append(tarea)
    print(f"Tarea '{tarea}' agregada.")

def eliminar_tarea(indice):
    try:
        tarea_eliminada = tareas.pop(indice - 1)
        print(f"Tarea '{tarea_eliminada}' eliminada.")
    except IndexError:
        print("Error: El número de tarea no es válido.")
    except ValueError:
        print("Entrada no válida. Por favor, introduce un número.")

def mostrar_tareas():
    if not tareas:
        print("No hay tareas pendientes.")
    else:
        print("\n--- TAREAS PENDIENTES ---")
        for i, tarea in enumerate(tareas, 1):
            print(f"{i}. {tarea}")
        print("-----------------------")
```

```python
while True:
    print("\n--- MENÚ DE TAREAS ---")
    print("1. Agregar tarea")
    print("2. Eliminar tarea")
    print("3. Mostrar tareas")
    print("4. Salir")
    print("-----------------------")

    opcion = input("Introduce tu opción: ")

    if opcion == '1':
        nueva_tarea = input("Introduce la nueva tarea: ")
        agregar_tarea(nueva_tarea)
    elif opcion == '2':
        mostrar_tareas()
        if tareas:
            try:
                num_tarea = int(input("Introduce el número de la tarea a eliminar: "))
                eliminar_tarea(num_tarea)
            except ValueError:
                print("Entrada no válida. Por favor, introduce un número entero.")
    elif opcion == '3':
        mostrar_tareas()
    elif opcion == '4':
        print("¡Hasta luego!")
        break
    else:
        print("Opción no válida. Por favor, intenta de nuevo.")
```
